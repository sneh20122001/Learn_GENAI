# What is a Vector Store?
- A vector store is a system or tool used to store and search vector embeddings.
- Embeddings are numerical representations of data (like text, images, audio, etc.) that capture semantic meaning.

- For example, "king" and "queen" are close in meaning, so their vectors will be close in space.

    - So a vector store:

        - Stores these vectors.

        - Allows fast similarity search (like: ‚ÄúFind me texts similar to this one‚Äù).

# Why Use a Vector Store?
We use vector stores for tasks where semantic search or machine learning is involved:

    - Semantic search: Find related documents even if the words don't match exactly.

    - RAG (Retrieval-Augmented Generation): In AI apps like ChatGPT + your documents, you retrieve relevant context from a vector store.

    - Recommendations: Suggest items based on user preferences.

    - Clustering & classification of data based on meaning.

üß† Traditional keyword search can‚Äôt understand meaning ‚Äî vector stores can.

# Vector Store vs. Vector Database

| Feature              | **Vector Store**                                | **Vector Database**                                  |
|----------------------|--------------------------------------------------|------------------------------------------------------|
| **Definition**       | A tool/library for storing and searching embeddings (vectors) | A database system designed to manage and query vector data |
| **Complexity**       | Lightweight, minimal features                    | Full-fledged DBMS with rich features                 |
| **Persistence**      | Often in-memory only (can be temporary)          | Persistent, scalable storage                         |
| **Filtering**        | Not always supported                             | Supports metadata filtering, hybrid search           |
| **Scalability**      | Not easily scalable                              | Designed for horizontal scaling                      |
| **APIs**             | Usually low-level (Python, C++)                  | REST, GraphQL, gRPC APIs                             |
| **Examples**         | FAISS, Annoy, HNSWLib                            | Pinecone, Weaviate, Qdrant, Milvus                   |
| **Use Case**         | Local apps, prototyping, academic work           | Production-grade systems, real-time applications     |

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.vectorstores import Chroma

embeddings = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004")
```

```python
from langchain.schema import Document

# Create LangChain documents for IPL players

doc1 = Document(
        page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style and fitness, he has led the Royal Challengers Bangalore in multiple seasons.",
        metadata={"team": "Royal Challengers Bangalore"}
    )
doc2 = Document(
        page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his calm demeanor and ability to play big innings under pressure.",
        metadata={"team": "Mumbai Indians"}
    )
doc3 = Document(
        page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills, wicketkeeping, and leadership are legendary.",
        metadata={"team": "Chennai Super Kings"}
    )
doc4 = Document(
        page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his yorkers and death-over expertise.",
        metadata={"team": "Mumbai Indians"}
    )
doc5 = Document(
        page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his quick fielding and match-winning performances make him a key player.",
        metadata={"team": "Chennai Super Kings"}
    )
```

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    embedding_function=embeddings,
    persist_directory='my_chroma_db',
    collection_name='sample'
)
```

```python
# add documents
vector_store.add_documents(docs)
```

```python
# view documents
vector_store.get(include=['embeddings','documents', 'metadatas'])
```

```python
# search documents
vector_store.similarity_search(
    query='Who among these are a bowler?',
    k=2
)
```

```python
# search with similarity score
vector_store.similarity_search_with_score(
    query='Who among these are a bowler?',
    k=2
)
```

```python
# meta-data filtering
vector_store.similarity_search_with_score(
    query="",
    filter={"team": "Chennai Super Kings"}
)
```
